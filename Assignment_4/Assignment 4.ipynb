```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
data=np.genfromtxt('linear-data-set-for-regression.csv',delimiter=',')
```

```python
x=data[:,1]
y=data[:,0]
```

```python

def cost(x,y,w,b):
    m=x.shape[0]
    cost=0
    
    
    for i in range(m):
        f=w*x+b
        cost=cost+(f-y)**2
    
    total_cost=cost/(2*m)
    
    return total_cost
```

```python
def compute_grad(x,y,w,b):
    m=x.shape[0]
    dj_dw=0
    dj_db=0
    
    for i in range(m):
        f=w*x[i]+b
        dj_dw_i=(f-y[i])*x[i]
        dj_db_i=(f-y[i])
        dj_dw=dj_dw+dj_dw_i
        dj_db=dj_db+dj_db_i
        
    dj_db1=dj_dw/m
    dj_db0=dj_db/m
    
    return dj_db0,dj_db1
```

```python
def gradient_descent(x,y,w_in,b_in,n,iterations,cost,compute_grad):
    J_his=[]
    p_his=[]
    
    b=b_in
    w=w_in
    
    for i in range(iterations):
        #Calculating the Derivative terms
        dj_db,dj_dw=compute_grad(x,y,w,b)
        
        #Updating the parameters
        w=w-n*dj_dw
        b=b-n*dj_db
        
        
        
        if i<1000000:
            J_his.append(cost(x,y,w,b))
            p_his.append([w,b])
        #     if i% math.ceil(iterations/10) == 0:
        #       print(f"Iteration {i:4}: Cost {J_his[-1]:0.2e} ",
        #           f"dj_db1: {dj_dw: 0.3e}, dj_db0: {dj_db: 0.3e}  ",
        #           f"b1: {w: 0.3e}, b0:{b: 0.5e}")
    return w,b,J_his,p_his
```

```python
#Parameters 
slope=int(input("Enter the value of initial slope: "))
intercept=int(input("Enter the value of the initial intercept: "))
n=float(input("Enter the value of n: "))
z=int(input("Enter the number of iterations: "))

slope_final,intercept_final,J_history,p_history=gradient_descent(x,y,slope,intercept,n,z,cost,compute_grad)

print(f"(Slope ,Intercept) found by gradient descent: ({slope_final:8.4f},{intercept_final:8.4f})")
```

```python
plt.plot(range(z), J_history, marker='o', linestyle='-', color='b')
plt.xlabel('Iteration Number')
plt.ylabel('Cost Function Value')
plt.title('Cost Function Value vs. Iteration Number')
plt.grid(True)
plt.show()
```
